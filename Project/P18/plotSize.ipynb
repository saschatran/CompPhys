```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

dfSArr = pd.read_csv("sumArr.csv")
dfSMat = pd.read_csv("sumMat.csv")
```

```python

df = dfSArr
df["Matrix Time"] = dfSMat["Matrix Time"]

indices = df.loc[df["Size of array"] == " ---"].index

dfSAll = []

pre = 0
for index in indices:
    dfSAll.append(df.iloc[pre:index])
    pre = index + 1

dfSAll[5]
```

```python
from matplotlib import ticker
def plotLine(data, counter):
    plt.clf()
    
    d = data[counter]
    ax = d.plot(x='Size of array', y='Matrix Time')
    ax = d.plot(x='Size of array', y='Array Time', ax=ax)

    init = int(data[counter]["Size of array"].iloc[0]) - 1
    
    ax.yaxis.set_major_formatter(plt.ScalarFormatter())
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

    ax.xaxis.set_major_locator(ticker.MultipleLocator(100))

    
    plt.xticks(rotation=45)
    plt.legend(loc='upper left')

    plt.savefig(f"len{init}.png")
    

        
```

```python
plotLine(dfSAll, 0)
```

```python
plotLine(dfSAll, 1)
```

```python
plotLine(dfSAll, 2)
```

```python
plotLine(dfSAll, 3)
```

```python
plotLine(dfSAll, 4)
```

```python
plotLine(dfSAll, 5)
```
