```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dfA = pd.read_csv("mat.csv")
```

```python
from matplotlib.colors import ListedColormap
from matplotlib.colors import LogNorm
import matplotlib.ticker as ticker

# Create the 'magma_no_black' colormap
df = dfA.groupby(["initL", "initM"], as_index=False).mean()
df_pivot = df.pivot(index='initL', columns='initM', values='Matrix Time')
cmap = sns.color_palette("magma_r",  as_cmap=True)
cmap.set_bad('#D3D3D3')

ax = sns.heatmap(df_pivot, linewidths=0.5, linecolor="grey", cmap=cmap, norm=LogNorm())

ax.invert_yaxis()

plt.title("Performance with different matrix sizes")
plt.savefig("ml.png")
```

```python
df = dfA.groupby(["initL", "initM"], as_index=False).mean()
df_pivot = df.pivot(index='initL', columns='initM', values='Count')
cmap = sns.color_palette("magma_r",  as_cmap=True)
cmap.set_bad('#D3D3D3')
ax = sns.heatmap(df_pivot, linewidths=0.5, linecolor="grey", cmap=cmap, norm=LogNorm())

ax.invert_yaxis()
plt.title("Total number of binary searches with different matrix sizes")
plt.savefig("bs.png")
```
