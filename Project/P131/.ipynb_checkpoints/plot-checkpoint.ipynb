```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("verlet.csv")
```

```python
def plotLine(df, yData):
    df.plot(x="Time", y=yData)
    plt.ylabel(yData)
    plt.show()
```

```python
plotLine(df, "Position")
```

```python
plotLine(df, "Velocity")
```

```python
hot_time = []
cold_time = []
hot_temp = []
cold_temp = []

with open("MDiter_cold",'r') as f:
    cold = f.readlines()

with open("MDiter_hot",'r') as f:
    hot = f.readlines()
     
for line in cold:
    cold_time.append(float(line.split()[0]))
    cold_temp.append(float(line.split()[4]))
    
for line in hot:
    hot_time.append(float(line.split()[0]))
    hot_temp.append(float(line.split()[4]))
    

    
equi_time = cold_time[int(len(cold_time) / 2):] + hot_time[int(len(hot_time) / 2):]
equi_temp = cold_temp[int(len(cold_temp) / 2):] + hot_temp[int(len(hot_temp) / 2):]
    
fig, ax = plt.subplots()

ax.scatter(cold_time, cold_temp, s=0.5, alpha=0.5, c='darkblue')
ax.scatter(hot_time, hot_temp, s=0.5, alpha=0.5, c='darkred')
ax.set(xlabel='MD step (0.1s)', ylabel='Temperature (K)',
       title='Temperature evolution over time')
ax.legend(['cold.dat', 'hot.dat'])
ax.set_xticks(range(0, 200001, 50000))
z = np.polyfit(equi_time, equi_temp, 2)
p = np.poly1d(z)
plt.plot(cold_time, p(cold_time))


fig.savefig("final_plot.png", dpi=300)
plt.show()
```
